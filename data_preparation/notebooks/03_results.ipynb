```python
import pandas as pd
from pathlib import Path
import geopandas as gpd

print('Libraries are downloaded')
```

```python
PROJECT_ROOT = Path("..").resolve().parent

PROCESSED = PROJECT_ROOT / "data" / "processed"
```

```python
df_total_pop = pd.read_csv(PROCESSED / "MD2_share_65_plus.csv")
df_total_pop.head()
```

```python
df = pd.read_csv(PROCESSED / "MD5_age_houses_occupation.csv")
df.head()
```

```python
df_disp = pd.read_csv(PROCESSED / "MD4_dispertion_places.csv")
df_disp.head()
```

```python
total_65plus = df_total_pop["pop_65plus"].sum()
total_pop    = df_total_pop["tot_pop"].sum()

italy_share_65plus = total_65plus / total_pop * 100

print(f"Share of residents aged 65+ in Italy (2025): {italy_share_65plus:.2f}%")
```

```python
# K — number Old & Empty regions
mask_old_empty = (df["high_65"] == 1) & (df["high_vac"] == 1)
K = df[mask_old_empty].shape[0]

N = df.shape[0]

print("K =", K)
print(f"Tile title: {K} / {N}")
```

```python
idx_max = df_disp["dispersed_index"].idxmax()
most_dispered_row = df_disp.loc[idx_max]

most_dispered_region = most_dispered_row["region"]  
most_dispered_value = most_dispered_row["dispersed_index"]

print("Most dispersed region:", most_dispered_region)
print("Dispersed Settlements Index:", most_dispered_value)
```

```python
mean_disp = df_disp["dispersed_index"].mean()
ratio = most_dispered_value / mean_disp

print(f"National average index: {mean_disp:.2f}")
print(f"Top region vs average: {ratio:.2f}×")
```
